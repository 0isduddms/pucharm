```python
from collections import defaultdict

description = \
'Artificial intelligence (AI, also machine intelligence, MI) is ' +\
'intelligence exhibited by machines, rather than ' +\
"humans or other animals (natural intelligence, NI)."

char_freq = defaultdict(int)

for i in description :
    char_freq[i] += 1
    
print(sorted(char_freq.items(), key=lambda x : x[1],reverse=True)[:10])
```

```python
from collections import defaultdict

price = [
    ('apple',50),
    ('banana',120),
    ('grape',500),
    ('apple',70),
    ('lemon',150),
    ('grape',1000)
]

d= defaultdict(list)

for key, value in price :
    d[key].append(value)

print(d)
print([sum(x) / len(x) for x in d.values()])
```

```python
from collections import Counter

description = \
'Artificial intelligence (AI, also machine intelligence, MI) is ' +\
'intelligence exhibited by machines, rather than ' +\
"humans or other animals (natural intelligence, NI)."

char_freq = Counter(description)

print(char_freq.most_common(10))
```

```python
from collections import defaultdict
from collections import Counter

a=[3,2,1,4,5,5,3,5,1]

d= defaultdict(int)

for key in a :
    d[key] += 1
    
c=Counter(a)

print(d)
print(c)
```

```python
import pandas as pd
from pandas import DataFrame

dupli_data = DataFrame({'col1' : [1,2,2,3,3,4,4,5,6,7,8],
                       "col2" : ['a','b','b','c','d','e','e','e','e','e','e']})
# ë¶„í•  ë¦¬ìŠ¤íŠ¸ ìƒì„±
cut_list=[1,3,5,8]

# êµ¬ê°„ ì´ë¦„ ë¦¬ìŠ¤íŠ¸ ìƒì„±
group_names=['first','second','third']

# êµ¬ê°„ ë¶„í• 
# dupli_dataì˜ col1 ì¹¼ëŸ¼ì„ cut_list ëŒ€ë¡œ ë¶„í• í•¨
# group_namesì„ ì´ìš©í•˜ì—¬ êµ¬ê°„ ì´ë¦„ ë¶™íž˜
cut_data = pd.cut(dupli_data.col1,cut_list,labels=group_names)

# ê° êµ¬ê°„ì˜ ìˆ˜ ì§‘ê³„
pd.value_counts(cut_data)
```
