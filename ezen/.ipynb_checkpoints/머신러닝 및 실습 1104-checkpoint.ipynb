```python
from collections import defaultdict

description = \
'Artificial intelligence (AI, also machine intelligence, MI) is ' +\
'intelligence exhibited by machines, rather than ' +\
"humans or other animals (natural intelligence, NI)."

char_freq = defaultdict(int)

for i in description :
    char_freq[i] += 1
    
print(sorted(char_freq.items(), key=lambda x : x[1],reverse=True)[:10])
```

```python
from collections import defaultdict

price = [
    ('apple',50),
    ('banana',120),
    ('grape',500),
    ('apple',70),
    ('lemon',150),
    ('grape',1000)
]

d= defaultdict(list)

for key, value in price :
    d[key].append(value)

print(d)
print([sum(x) / len(x) for x in d.values()])
```

```python
from collections import Counter

description = \
'Artificial intelligence (AI, also machine intelligence, MI) is ' +\
'intelligence exhibited by machines, rather than ' +\
"humans or other animals (natural intelligence, NI)."

char_freq = Counter(description)

print(char_freq.most_common(10))
```

```python
from collections import defaultdict
from collections import Counter

a=[3,2,1,4,5,5,3,5,1]

d= defaultdict(int)

for key in a :
    d[key] += 1
    
c=Counter(a)

print(d)
print(c)
```

```python
import pandas as pd
from pandas import DataFrame

dupli_data = DataFrame({'col1' : [1,2,2,3,3,4,4],
                       "col2" : ['a','b','b','c','d','e','e']})

col_map = { 'a' : 'A',
          'b' : 'B',
          'c' : 'C',
          'd' : 'D',
          'e' : 'E'}

# 새로운 컬럼 col3을 추가 해당 데이터가 없는 경우는 NaN
# dupli_data의 col2 칼럼을 기반으로 대응
dupli_data['col3'] = dupli_data['col2'].map(col_map)
dupli_data
```

```python
import pandas as pd
from pandas import DataFrame

dupli_data = DataFrame({'col1' : [1,2,2,3,3,4,4,5,6,7,8],
                       "col2" : ['a','b','b','c','d','e','e','e','e','e','e']})
# 분할 리스트 생성
cut_list=[1,3,5,8]

# 구간 이름 리스트 생성
group_names=['first','second','third']

# 구간 분할
# dupli_data의 col1 칼럼을 cut_list 대로 분할함
# group_names을 이용하여 구간 이름 붙힘
cut_data = pd.cut(dupli_data.col1,cut_list,labels=group_names)

# 각 구간의 수 집계
pd.value_counts(cut_data)
```
